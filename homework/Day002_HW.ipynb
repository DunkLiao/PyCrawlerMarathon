### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
import csv

# 開啟 CSV 檔案
with open('./data/example.csv', newline='',encoding="utf-8") as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row)
```

```python
# 1. 取出班次一的每一個時間
col_name = '班次1'
# 開啟 CSV 檔案
with open('./data/example.csv', newline='',encoding="utf-8") as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.DictReader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row[col_name])
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存
col_name = '班次1'
my_list=[]
# 開啟 CSV 檔案
with open('./data/example.csv', newline='',encoding="utf-8") as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.DictReader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        my_list.append(row[col_name])

print(len(my_list)) #列出長度
my_list #顯示內容
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存
dict_time={}
col_name_list = ['班次1','班次2','班次3','班次4','班次5']

# 開啟 CSV 檔案
i=0
with open('./data/example.csv', newline='',encoding="utf-8") as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.DictReader(csvfile)
    # 以迴圈輸出每一列    
    for row in rows:
        i= i+1
        for col_name in col_name_list:                
            dict_time[col_name+'_'+str(i)] = row[col_name]
        
print(len(dict_time)) #列出長度
dict_time #顯示內容
```
