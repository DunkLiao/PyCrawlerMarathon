# API 資料串接 - 以 Dcard API 實作範例

* 了解 Dcard API 使用方式與回傳內容
* 撰寫程式存取 API 且解析 JSON 格式資料


## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

#### 讀取資料


```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

```python
import json
data = json.loads(response)
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
print('筆數:%d'% int(len(data)))
print('包含欄位:%s' % data[0].keys())    
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
col_name = {'標題':'title','貼文時間':'createdAt','留言人數':'commentCount','按讚人數':'likeCount'}
for d in data:
    for k in col_name:        
        print('{0} = {1}'.format(k,d[col_name[k]]))
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
commentCount = 0
likeCount = 0
for d in data:
    commentCount += d['commentCount']
    likeCount += d['likeCount']

print('熱門文章的「平均留言人數」與「平均按讚人數」')
print(commentCount/len(data), likeCount/len(data))

r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=false')
response = r.text
data = json.loads(response)
commentCount = 0
likeCount = 0
for d in data:
    commentCount += d['commentCount']
    likeCount += d['likeCount']

print('非熱門文章的「平均留言人數」與「平均按讚人數」')
print(commentCount/len(data), likeCount/len(data))
```
