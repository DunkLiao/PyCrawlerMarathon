```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

```python
import json
data = json.loads(response)
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
print('ç­†æ•¸:%d'% int(len(data)))
print('åŒ…å«æ¬„ä½:%s' % data[0].keys())    
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
col_name = {'æ¨™é¡Œ':'title','è²¼æ–‡æ™‚é–“':'createdAt','ç•™è¨€äººæ•¸':'commentCount','æŒ‰è®šäººæ•¸':'likeCount'}
for d in data:
    for k in col_name:        
        print('{0} = {1}'.format(k,d[col_name[k]]))
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
commentCount = 0
likeCount = 0
for d in data:
    commentCount += d['commentCount']
    likeCount += d['likeCount']

print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€')
print(commentCount/len(data), likeCount/len(data))

r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=false')
response = r.text
data = json.loads(response)
commentCount = 0
likeCount = 0
for d in data:
    commentCount += d['commentCount']
    likeCount += d['likeCount']

print('éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€')
print(commentCount/len(data), likeCount/len(data))
```
