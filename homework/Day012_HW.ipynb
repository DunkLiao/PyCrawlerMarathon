####  讀取來源網址

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 取出今天所有的發文

```python
import datetime
import re

today = datetime.date.today()
today_pattern = re.compile('{0}\w*'.format(today.strftime("%Y/%m/%d")))

for d in soup.find(class_="part_list_2").find_all('h3'):
    date_display = d.find(class_="date").text
    title = d.find_all('a')[-1].text
    if today_pattern.match(date_display):
        print(date_display,title)
```
