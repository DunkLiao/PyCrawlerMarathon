```python
url = "https://cat-fact.herokuapp.com/facts"
r = requests.get('https://api.github.com/events')
r.text
```

```python
js_obj = json.loads(r.text)
```

```python
import pprint
pprint.pprint(js_obj)
```

```python
#列舉物件
for js in js_obj:
    print('id=%s' % js['id'])
```

```python
#寫入檔案資料
target_file = './data/data.json'
with open(target_file, 'w') as f:
    json.dump(js_obj, f)
    
#讀取資料
with open(target_file, 'r') as f:
    data3 = json.load(f)
    print(data3)    
```
